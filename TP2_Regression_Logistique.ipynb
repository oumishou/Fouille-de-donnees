```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from mlxtend.plotting import plot_decision_regions  # Importer la fonction plot_decision_regions

data = load_breast_cancer()
# Utiliser uniquement les deux premires fonctionnalits pour une visualisation bidimensionnelle
X = data.data[:, :2]  
y = data.target

# Diviser les donnes en ensembles d'entranement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Initialiser et entraner le modle de rgression logistique
model = LogisticRegression()
model.fit(X_train, y_train)

# Tracer la courbe de dcision
#Taille de 10 pouces de largeur sur 7 pouces de hauteur.
plt.figure(figsize=(10, 7))
plot_decision_regions(X_train, y_train, clf=model, legend=2)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[8])
plt.title('Regression Logistique pour les donnes cancer du sein')
plt.show()
```
