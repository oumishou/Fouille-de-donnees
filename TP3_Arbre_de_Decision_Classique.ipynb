```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier  # Importer DecisionTreeClassifier
from sklearn import tree  # Pour la visualisation de l'arbre de décision

data = load_breast_cancer()
# Utiliser uniquement les deux premières fonctionnalités pour une visualisation bidimensionnelle
X = data.data[:, :2]  
y = data.target

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

# Initialiser et entraîner le modèle d'arbre de décision
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)

# Tracer l'arbre de décision
plt.figure(figsize=(15, 10))  # Ajout de la parenthèse fermante ici
tree.plot_tree(tree_model, feature_names=data.feature_names[:2], class_names=data.target_names, filled=True)
plt.title('Arbre de Décisions pour les données sur le cancer du sein')
plt.show()
```
