```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Chargement des données d'entraînement depuis le fichier CSV
train_data = pd.read_csv("train.csv")

# Sélection de la colonne 'SalePrice' comme cible
target_column = 'SalePrice'
print(train_data)
# Séparation des caractéristiques (features) et de la cible (target) pour les données d'entraînement
X_train = train_data.drop(columns=[target_column])
y_train = train_data[target_column]

# Convertir les variables catégoriques en variables indicatrices
X_train = pd.get_dummies(X_train)

# Création du modèle d'arbre de décision regressor
model = DecisionTreeRegressor(max_depth=8, min_samples_leaf=0.08, random_state=4)

# Entraînement du modèle sur les données d'entraînement
model.fit(X_train, y_train)

# Affichage de l'arbre de décision
plt.figure(figsize=(20,15))
plot_tree(model, filled=True, feature_names=X_train.columns)
plt.show()
```
