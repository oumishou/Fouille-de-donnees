```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

# Chargement des donnes d'entranement depuis le fichier CSV
train_data = pd.read_csv("train.csv")

# Slection de la colonne 'SalePrice' comme cible
target_column = 'SalePrice'
print(train_data)
# Sparation des caractristiques (features) et de la cible (target) pour les donnes d'entranement
X_train = train_data.drop(columns=[target_column])
y_train = train_data[target_column]

# Convertir les variables catgoriques en variables indicatrices
X_train = pd.get_dummies(X_train)

# Cration du modle d'arbre de dcision regressor
model = DecisionTreeRegressor(max_depth=8, min_samples_leaf=0.08, random_state=4)

# Entranement du modle sur les donnes d'entranement
model.fit(X_train, y_train)

# Affichage de l'arbre de dcision
plt.figure(figsize=(20,15))
plot_tree(model, filled=True, feature_names=X_train.columns)
plt.show()
```
